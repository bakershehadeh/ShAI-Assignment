# Descriptive Statistics

```python
# To illustrate the statistics of the numeric features, we need to include those informative features
# and remove the ones that do not imply any infromation like (Id)
# Remove the following features: Id, Year, Notes and Status

drop_cols = ['Id','Year','Notes','Status']
Numeric_features = df.drop(drop_cols, axis=1)
Numeric_features.describe()
```

```python
# To estimate the statistics of the salaries (TotalPay)
Numeric_features['TotalPay'].describe()
```

```python
print("The mean of the salaries is")
print(Numeric_features['TotalPay'].mean())
print("------------------------------")
print("The median of the salaries is")
print(Numeric_features['TotalPay'].median())
print("------------------------------")
print("The mode of the salaries is")
print(Numeric_features['TotalPay'].mode().iloc[0])
print("------------------------------")
print("The maximum salary is")
print(Numeric_features['TotalPay'].max())
print("------------------------------")
print("The minimum salary is") # There's a negative number! Does it consider an outlier?
print(Numeric_features['TotalPay'].min())
print("------------------------------")
print("The range of salaries is")
print(Numeric_features['TotalPay'].max() - Numeric_features['TotalPay'].min())
print("------------------------------")
```

# Data Cleaning

```python
print(df.isnull().sum())
```

```python
# There are six features that contain missing values(NaN),
# Several methods will be used to handle the missing values depending on the situation
# For instance, The features (Notes and Status) will be removed from the dataset since all of the rows
# don't conatin any value

drop_cols = ['Notes','Status']
df_cleaned =  df.drop(drop_cols, axis=1)

# Check the characteristic of the (BasePay)

print(df_cleaned['BasePay'].max())
print(df_cleaned['BasePay'].min())
print(df_cleaned['BasePay'].mean())

```

```python
# Extract the records that contain NaN in the BasePay to decide the convenient procedure
df_BasePay = df_cleaned[df_cleaned['BasePay'].isnull()]
df_BasePay.tail()
```

```python
# I decide to fill the null values with Zeros, I assume that those employees work based on commission so they don't have base salary
# but have other pay. Also the benefits will be zeros for the null according to the same logic

df_cleaned.fillna(0, inplace=True)
df_cleaned.isnull().sum()
```

# Basic Data Visualization

```python
plt.figure(figsize=(8, 4))
plt.hist(df_cleaned['TotalPay'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Salaries')
plt.xlabel('Salaries')
plt.ylabel('Frequency')
plt.show()
```

# Grouped Analysis

```python
# Gruoping using the employee name
drop_cols = ['Id','Year']
df_cleaned_numeric = df_cleaned.drop(drop_cols, axis=1)

df_grouped = df_cleaned_numeric.groupby('EmployeeName')

# Applying statistical operations to each group
result = df_grouped.agg(['mean', 'median', 'min', 'max'])
```

# Simple Correlation Analysis

```python
# The follwoing figures show the correlation between the total salary and the base one

plt.figure(figsize = (10,10))
sns.heatmap(df_cleaned.corr().loc[['TotalPay'], ['BasePay']], square=True, annot=True, cmap='viridis',
            linewidths=.5)
plt.show()
```
